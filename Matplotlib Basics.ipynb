```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,5)
```

```python
y = x**3
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Functional method plot")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"b")

plt.subplot(1,2,2)
plt.plot(y,x,"r")

plt.show()
```

```python
#Object oriented method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Object Oriented method plotting")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) #Empty space for outer plot
axes2 = fig.add_axes([0.19,0.45,0.45,0.45]) #Empty space for inner plot
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) #Empty space for outer plot
axes2 = fig.add_axes([0.19,0.45,0.45,0.45]) #Empty space for inner plot


axes1.plot(x,y,"r")
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("Outer Plot")

axes2.plot(y,x)
axes2.set_xlabel("X axis")
axes2.set_ylabel("Y axis")
axes2.set_title("Inner Plot")
```
